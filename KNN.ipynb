```python
import numpy as np
import pandas as pd
import math
from scipy.stats import mode
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("knnDataset.csv")
df.head(10)
```

```python
df.head(20)
```

```python
data = np.array(df)
print(data)
```

```python
Y = data[:, 1:2]
print(Y.shape)
X = data[:, 2:10]
print(X.shape)
```

```python
for i in range(len(Y)):
    if(Y[i] == 'M'):
        Y[i] = 1
    else:
        Y[i] = 0
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
def euclidian(p1, p2):
    dist = math.sqrt(np.sum((p1 - p2) ** 2))
    return dist
```

```python
def knn(x, y, test_input, k):
    output = []
    for i in test_input:
        dist_arr = []
        for j in x:
            dist = euclidian(i, j)
            dist_arr.append(dist)
        dist_arr = np.array(dist_arr)
            
        arr = np.argsort(dist_arr)[:k]
        labels = y[arr]
        a = mode(labels)
        a = a.mode[0]
        output.append(a)
    
    output = np.array(output)
    return output
```

```python
y_pred = knn(x_train, y_train, x_test, 5)
c = 0
for i in range(len(y_pred)):
    if(y_pred[i][0] == y_test[i][0]):
        c = c + 1;
        
print(c/len(y_pred))
```
